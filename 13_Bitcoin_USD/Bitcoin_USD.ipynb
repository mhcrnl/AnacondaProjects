```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas_datareader.data as web
import datetime
```

```python
start_date = datetime.datetime(2013,4,28)
end_date = datetime.datetime.today()
```

```python
bitcoin = web.DataReader('BTC-USD', 'yahoo', start_date, end_date)
bitcoin.tail()

```

```python
bitcoin['Open'].plot(label='BTC-USD', figsize=(16,8), title='Opening Prices')
plt.legend()
```

```python
bitcoin['Volume'].describe()
```

```python
bitcoin['Volume'].plot(label='BTC_USD', figsize=(16,8), title='Volume Traded')
plt.legend()
```
